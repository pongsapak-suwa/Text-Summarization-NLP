# 2.  การทำ Word Tokenize

---
นำข้อความมาตัดออกมาเป็นคำ เพื่อจะนำไปนับความถี่ของคำภายในบทความ


```python
word_th = word_tokenize(text,engine='newmm') #การนำข้อความมาตัดออกให้เป็นคำ
word_th
```

# 3. นับความถี่ของคำเก็บในไว้ใน Dictionary


---


*   Text Cleaning ตัดคำที่เป็นเครื่องหมายวรรคตอน stopwords และสัญลักษณ์ต่างๆ
*   ยิ่งมีความถี่มากๆหมายความว่ามีความสำคัญมาก




```python
word_freq_th = {} #1. สร้าง Dictionary
for word in word_th: #2. เช็คคำที่ได้จากการตัดคำว่าต่างกับเงื่อนไขด้านล่างนี้หรือไม่ ถ้าตรงให้ค่าเป็น 1 ถือว่าเป็นคำเเรกที่เจอ เเต่ถ้าไม่ตรงให้บวกค่าเพิ่มเป็น 1 ถือว่าเป็นคำที่เคยเจอเเล้ว โดยเงื่อนไขที่ว่ามีดังนี้
  if word not in thai_stopwords: #เงื่อนไขที่ 1: มีคำที่เป็นคำในลิสต์ของคำ stopwords หรือไม่ ถ้าไม่ใช่ไปเช็คต่อที่เงื่อนไขที่ 2 
    if word not in punctuation:  #เงื่อนไขที่ 2: มีเครื่องหมายวรรคตอนหรือไม่ !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~ ถ้าไม่ใช่ไปเช็คต่อที่เงื่อนไขที่ 3 
      if word not in " ": #เงื่อนไขที่ 3: มี white space หรือใหม่ ถ้าไม่ใช่ไปเช็คต่อที่เงื่อนไขที่ 4 
        if word not in word_freq_th.keys(): #เงื่อนไขที่ 4: word นี้อยู่ใน key ของ dictionary word_freq_th ที่เราใช้เก็บค่าความถี่หรือไม่ถ้าไม่อยู่ ถือว่าตรงกับเงื่อนไขทั้งหมดให้เพิ่มค่าเป็น 1 ถ้าอยู่ถือว่าไม่ตรงให้ไปที่ else เเล้วบวกค่าเพิ่ม 1 ค่า 
          word_freq_th[word] = 1 #เพิ่มค่าเป็น 1 เนื่องจากตรงกับทั้ง 4 เงื่อนไขที่ผ่านมา
        else:
          word_freq_th[word] += 1 #บวกค่าเพิ่มอีก 1 ค่า
```


**Sorted เพื่อเรียงลำดับความถี่จากมากไปน้อย**

---


```python
sorted(word_freq_th.items(), key=lambda x: x[1],reverse=True) #เเสดงค่าความถี่ที่นับได้โดยเรียงจากหลังไปหน้า (มากไปน้อย) ด้วยฟังก์ชัน sorted() เเละให้ค่า parameter reverse=True เเละ "key=lambda x: x[1]" เป็นการกำหนดว่าเรียงโดยใช้ตัว value
```

# 4. การทำ Normalize ปรับค่าให้อยู่ในช่วง 0 ถึง 1

---


ทำเพื่อให้ช่วงของข้อมูลอยู่ในช่วงใกล้เคียงกันเพื่อลดความซ้ำซ้อนของข้อมูลและทำให้ประสิทธภาพดีขึ้น



*   สูตร Word_freq_th[word]=ค่าความถี่ของแต่ละคำ/ค่าความถี่สูงสุด

```python
#ทำการ normalize ความถี่ที่ได้เพื่อปรับช่วงของข้อมูลให้อยู่ในช่วงใกล้เคียงกัน
max_freq_th = max(word_freq_th.values())
for word in word_freq_th.keys():
  word_freq_th[word] = word_freq_th[word]/max_freq_th
print(word_freq_th)
```

```python
sorted(word_freq_th.items(), key=lambda x: x[1],reverse=True)
```

#5. การทำ Sentence Tokenize

---

การแบ่งข้อความออกเป็นประโยคย่อยๆทำให้วิเคราะห์ข้อความได้ดีขึ้น

ฟังก์ชั่น sent_tokenize มากจากการ import pythainlp
* ใช้ในการตัดข้อความออกมาเป็นประโยค
* crfcut model คือ การแบ่งประโยคโดยใช้ Conditional Random Fields (CRF)ที่ผ่านการฝึกอบรมในชุดข้อมูล TED
* Conditional Random Fields (CRF) เป็น sequence model ที่ได้รับความนิยมมากที่สุดเนื่องจากทำงานได้ดี train ได้โดยใช้เวลาไม่มาก ไม่ต้อง tune hyperparamters ให้ยุ่งยาก

Performance: - ORCHID - space-correct accuracy 87% vs 95% state-of-the-art

[อ้างอิง1](https://attapol.github.io/compling/information_extraction.html)
[อ้างอิง2](https://pythainlp.github.io/docs/2.3/api/tokenize.html#module-pythainlp.tokenize.crfcut)

```python
sent_th = sent_tokenize(text,engine="crfcut") #การนำข้อความมาตัดออกให้เป็นประโยค 
print(sent_th)
```

#6. คำนวณคะแนนของประโยคที่ตัดได้

---


โดยใช้ ค่าความถี่ของคำและประโยคที่ตัดได้

```python
sent_scores_th = {} #สร้าง dictionary 
for sent in sent_th: #นำประโยคที่ตัดไว้ทุกประโยคมาคำนวณ
  for word in sent: #เช็ค[คำ]ที่มีในประโยค A
    if word in word_freq_th.keys(): #ถ้าคำในประโยค A มีใน dictionary ของ word_freq_th(เก็บความถี่ของคำที่ตัดได้)
      if sent not in sent_scores_th.keys(): #ถ้าประโยคไม่ได้อยู่ใน dictionary ของ sent_scores_th
        sent_scores_th[sent] = word_freq_th[word] #ให้ sentence scores เท่ากับ ค่าความถี่ที่ normalize เเล้วของ word frequencies
      else:   #ถ้าประโยคอยู่ใน dictionary ของ sent_scores_th 
        sent_scores_th[sent] += word_freq_th[word] #ให้ บวกเพิ่ม sentence scores เท่ากับ ค่าความถี่ที่ normalize เเล้วของ word frequencies ไปอีก 1 ครั้ง
```

```python
print(sent_scores_th)
```


**sorted เพื่อหาประโยคที่มีความสำคัญมากที่สุดตามลำดับ:**

---

*  ประโยคที่ทำการ sorted แล้วได้มาตรงนี้จะเป็นประโยคที่นำไปสร้างสรุปบทความ

```python
sorted(sent_scores_th.items(), key=lambda x: x[1], reverse=True) #เเสดงค่าความถี่ที่นับได้โดยเรียงจากหลังไปหน้า (มากไปน้อย) ด้วยฟังก์ชัน sorted() เเละให้ค่า parameter reverse=True เเละ "key=lambda x: x[1]" เป็นการกำหนดว่าเรียงโดยใช้ตัว value 
```

# 7. สร้างสรุปบทความจากคะแนนของประโยคที่คำนวณได้

---



```python
from heapq import nlargest #หาเเถวที่มีค่าสูงสุด
```

 select_len_th เป็นตัวที่กำหนดว่าจะให้บทความสั้นหรือยาวมากแค่ไหน โดยค่าน้อยบทความจะสั้นลง ค่ามากบทความจะยาวขึ้น

```python
select_len_th = int(len(sent_scores_th)*0.5) # (max=1)ความยาวเเละเนื้อหาที่สรุปก็จะมากขึ้น ขึ้นอยู่ที่ว่าเราต้องการให้บทความมีความสั้นยาวสรุปได้กระชับมากเเค่ไหน
select_len_th
```

```python
sum_text = nlargest(select_len_th, sent_scores_th, key=sent_scores_th.get) #เเสดงประโยคที่มีความสำคัญมากที่สุดจากค่า sentence scores 
sum_text
```

```python
#รวมข้อมูลตัวที่ 1 เเละ 2 ที่ได้จาก dict ให้อยู่ใน paragraph เดียวกัน
sum_text = "".join(sum_text)
sum_text
```

```python
#ฟังก์ชันสำหรับเเสดงข้อความ 2 คอลัมน์
from textwrap import wrap
from itertools import zip_longest
def twoColumn(left, right, leftWidth=10, rightWidth=100, indent=1, separation=2):
    lefts = wrap(left, width=leftWidth)
    rights = wrap(right, width=rightWidth)
    results = []
    for l, r in zip_longest(lefts, rights, fillvalue=''):
       results.append('{0:{1}}{2:{5}}{0:{3}}{4}'.format('', indent, l, separation, r, leftWidth))
    return "\n".join(results)
```

```python
# เเสดงผลข้อมูล
print(f'---------------------------------------------ปรียบเทียบความยาวก่อนหลังสรุป---------------------------------------------\n')
print(f'ความยาวก่อนสรุปข่าว: {len(text)}')
print(f'ความยาวหลังสรุปข่าว (NLP): {len(sum_text)}')
print(f'------------------------------------------------รายละเอียดก่อนหลังสรุป----------------------------------------------------\n')
after_nlp = "After NLP: "
before = "Before: "
print(twoColumn(before,text))
print("\n")
print(twoColumn(after_nlp,sum_text))
```
